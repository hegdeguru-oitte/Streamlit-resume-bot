```python
# import cx_Oracle
# import pandas as pd
# import cx_Oracle as orc 
# cx_Oracle.init_oracle_client(lib_dir=r"C:\Users\gurhegde\Downloads\instantclient_21_13")
# # Establish connection to the Oracle database
# # connection= cx_Oracle.connect("grp_fin_ra_sb/Pppgg_ar148Pa@//hem1-scan:1525/eiwodsd")


user= 'GRP_FIN_RA_SB'
pwd = 'Telkom2207'
host = 'cdolsvwprep1' #could look like path or url depending on where it's hosted
service_name = 'SVWPRD1B.telkom.co.za'
portno = 1527 

# NGN
# user= 'GROUP_FIN_RA_READ'
# pwd = 'fin11ws3j'
# host = 'nxeim3-scan' #could look like path or url depending on where it's hosted
# service_name = 'eiwods_c3'
# portno = 1525 

connection = orc.connect(user, pwd, '{}:{}/{}'.format(host,portno,service_name))

cursor = connection.cursor()

# Read the list of service names from Excel file
data = pd.read_excel(r"C:\Users\gurhegde\OneDrive - Telkom\AG\test\Data2.xlsx")
service_names = data['SERVICENAME'].tolist()



# Add quotes around each service name
quoted_service_names = ["'{}'".format(name) for name in service_names]

# Split the list of service names into chunks of 1000
chunk_size = 1000
chunks = [quoted_service_names[i:i + chunk_size] for i in range(0, len(quoted_service_names), chunk_size)]

# Prepare the SQL query
# with open(r"C:\Users\gurhegde\OneDrive - Telkom\DB\Working\OMS_STATUS.sql", "r") as file:
#     sql_query = file.read()

# sql_query = "select * from DUMMYCBS_TEST where ND in ({})"
# sql_query = "select * from NTW_GH where CIRCUIT in ({})"
# sql_query = "select * from M_ICON_ALL_OPERATORS_FACT where SERVICE_NUMBER in ({})"
sql_query = "select * from service_history where SERVICE_NAME in ({})"

# Initialize an empty DataFrame to store the results
result = pd.DataFrame()

# Execute the SQL query for each chunk of service names
for chunk in chunks:
    placeholders = ','.join(chunk)
    query = sql_query.format(placeholders)
    cursor.execute(query)
    chunk_result = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    result = pd.concat([result, chunk_result], ignore_index=True)

print(result)

# Close cursor and connection
cursor.close()
connection.close()

# Save the results to Excel
result.to_excel(r"C:\Users\gurhegde\OneDrive - Telkom\AG\test\check8.xlsx", index=False)
```
